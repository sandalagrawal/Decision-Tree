```python
# Import some libraries
# import numpy package for arrays and stuff
import numpy as np
# import matplotlib.pyplot for plotting our result
import matplotlib.pyplot as plt
# import pandas for importing csv files
import pandas as pd
```

```python
# import dataset
# dataset = pd.read_csv('Data.csv')
# alternatively open up .csv file to read data
#We have created dataset by using numpy library in form of array
# we have 2-D array
dataset = np.array(
[['Asset Flip', 100, 1000],
['Text Based', 500, 3000],
['Visual Novel', 1500, 5000],
['2D Pixel Art', 3500, 8000],
['2D Vector Art', 5000, 6500],
['Strategy', 6000, 7000],
['First Person Shooter', 8000, 15000],
['Simulator', 9500, 20000],
['Racing', 12000, 21000],
['RPG', 14000, 25000],
['Sandbox', 15500, 27000],
['Open-World', 16500, 30000],
['MMOFPS', 25000, 52000],
['MMORPG', 30000, 80000]
])
# print the dataset
print(dataset)
```

```python
# select all rows by : and column 1
# by 1:2 representing features
X = dataset[:, 1:2].astype(int) #2nd column or 1st index column will be selected
# we are choosing independent variable and result will be stored into the X varia
# print X
print(X)
```

```python
# select all rows by : and column 2
# by 2 to Y representing labels
y = dataset[:, 2].astype(int) # we are selecting dependent variable value as a p
# print y
print(y)
```

```python
# Import library for splitting the dataset into train and test.
from sklearn.model_selection import train_test_split
#Using the train_test_split to create train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50)
```

```python
#Building the model
# Fit Simple Linear Regression to Training Data
from sklearn.linear_model import LinearRegression
regressor_lin = LinearRegression() #we are creating object of the linearregressio
regressor_lin.fit(X_train, y_train) # we are fitting the model by training data
```

```python
# import the regressor for decision tree
from sklearn.tree import DecisionTreeRegressor
# create a regressor object
regressor = DecisionTreeRegressor(random_state = 0)
# We are building the model as per training dataset
# fit the regressor with X and Y data
regressor.fit(X_train, y_train)
```

```python
y_pred_lin = regressor_lin.predict(X_test)
#prediction
y_pred = regressor.predict(X_test) #Prediting the model by testing data
```

```python
# predicting a new value
# test the output by changing values, like 3750
y_pred_lin = regressor_lin.predict([[9000]])
#y_pred = regressor.predict(3750)
y_pred = regressor.predict([[9000]])
# print the predicted price
#print("Predicted price: % d\n"% y_pred)
print("Predicted salary by Linear Regression: % d\n"% y_pred_lin)
print("Predicted salary: by Decision Tree % d\n"% y_pred)
# We are getting the price of a product by number of qty exactly close to the inp
```

```python
# We are fitting the model as per X and Y values
#Because it was fit by training data set
# We should take all values to print decision tree
# fit the regressor with X and Y data
regressor.fit(X, y)
regressor_lin.fit(X, y)
```

```python
# import export_graphviz
from sklearn.tree import export_graphviz
# export the decision tree to a tree.dot file
# for visualizing the plot easily anywhere
export_graphviz(regressor, out_file ='tree.dot',
feature_names =['Production Cost'])
```
